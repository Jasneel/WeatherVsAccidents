```python
#API
year_list = [2017,2018,2019]
population_171819 = []

for years in year_list:
    base_url = f'https://api.census.gov/data/{years}/pep/population'
    params = {
    'get' : 'POP',
    'for' : 'STATE:*',
    'key' : '60a95aa0242affd79076a8b964cd2c4516142c32'
    }
    data = requests.get(base_url,params=params).json()
    data = pd.DataFrame(data)
    data['YEAR'] = f'{years}'
    population_171819.append(data)
```

```python
pop_17 = population_171819[0]
pop_18 = population_171819[1]
pop_19 = population_171819[2]

ind_remove_10 = [0]
pop_17_list = pop_17.drop(ind_remove_10).rename(columns={
    0 : 'POPULATION',
    1 : 'STATE'
})

ind_remove_11 = [0]
pop_18_list = pop_18.drop(ind_remove_11).rename(columns={
    0 : 'POPULATION',
    1 : 'STATE'
})

ind_remove_12 = [0]
pop_19_list = pop_19.drop(ind_remove_12).rename(columns={
    0 : 'POPULATION',
    1 : 'STATE'
})

frames2 = [pop_17_list,pop_18_list,pop_19_list]
pop_171819 = pd.concat(frames2)
pop_171819
```

```python
pop_171819[["POPULATION"]] = pop_171819[["POPULATION"]].astype(int)
pop_171819.info()
```

```python
pop_171819['POPULATION'] = pop_171819['POPULATION'].astype(float)

by_state = pop_171819[(pop_171819["YEAR"] == "2019")]
by_state['POPULATION'] = by_state['POPULATION'] * 1.004
by_state["YEAR"] = 2020

by_state = pd.DataFrame(by_state)

frame3 = [pop_171819,by_state]
final_pop =  pd.concat(frame3)
final_pop
```

# Percentage of Number of Accidents in each year from 2017 to 2020

```python
df_weather_year = df.groupby('YEAR')['VE_TOTAL'].sum()
df_weather_year = pd.DataFrame(df_weather_year).sort_values('VE_TOTAL',ascending=False)
df_weather_year = df_weather_year.reset_index()
df_weather_year
plt.pie(df_weather_year['VE_TOTAL'],labels=df_weather_year['YEAR'],autopct='%1.1f%%',shadow=True)
plt.title('Accidents by Year')
plt.axis('equal')
plt.show()
plt.savefig("Accident_by_Year.png")
# In 2020 the pandemic started and it is the year with the most number of accidents. Stay at home rules were in effect.
```

The pie graph displays the number of accidents that occurred between 2017 and 2020. The pie chart demonstrates how closely correlated the accident percentages are for all 4 years. The year 2019 has the fewest accidents (24.4% of all accidents), while the year 2020 has the most accidents (25.8% of all accidents). which definitely caught us by surprise considering 2020 was also the year when the pandemic started and the regulations to stay at home were in effect.

# Accidents on different days of the Week

```python
day_country = df1.groupby(["DAY_WEEKNAME"])["VE_TOTAL"].sum()
day_country = pd.DataFrame(day_country).sort_values('VE_TOTAL',ascending=False).reset_index()
plt.scatter(day_country['DAY_WEEKNAME'],day_country['VE_TOTAL'])
plt.title('Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Accidents')
plt.savefig("Accident_by_Day.png")
```

We were able to determine which days of the week had the most and fewest accidents after conducting some analysis. Again, it was quite unexpected to learn that Saturdays, which are weekends, are when the majority of accidents occur. This disproved the theory that accidents would increase on weekdays as a result of everyone rushing to work and making wrong judgments while driving.

```python
saturday_df = df1.loc[(df1['DAY_WEEKNAME'] == 'Saturday'),:]
saturday_state = saturday_df.groupby('STATENAME')['VE_TOTAL'].sum()
saturday_state.sort_values(ascending=False)
```

# Numbers of Accidents relationship with differents times in a Day

```python
hour_country = df1.groupby(["HOURNAME"])["VE_TOTAL"].sum()
hour_country_df = pd.DataFrame(hour_country).sort_values('VE_TOTAL',ascending=False)
hour_country_df.plot.bar(figsize = (15, 6), rot = 90, fontsize = 15, xlabel= "Hours in a Day", ylabel = "Number of Accidents", title = "Numbers of Accidents in Different time of the Day")
plt.savefig("Accident_by_Time.png")
```

We carried out an in-depth analysis to identify the pattern that the number of accidents was rising as the day went on. According to our observations, accidents tended to happen more often between 6 and 7 p.m. and less frequently between 4 and 5 a.m.
